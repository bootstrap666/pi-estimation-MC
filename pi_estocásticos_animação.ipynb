```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

from matplotlib import animation, rc
from IPython.display import HTML
```

```python
# First set up the figure, the axis, and the plot element we want to animate
rect = patches.Rectangle((-1, -1), 2, 2, linewidth=1, edgecolor='black', facecolor='none')
circle = plt.Circle((0, 0), 1, color='gray')


fig, ax = plt.subplots()

ax.set_xlim((-1.1, 1.1))
ax.set_ylim((-1.1, 1.1))

ax.add_patch(circle)
ax.add_patch(rect)

line, = ax.plot([], [], lw=0, marker='x')

xlist = []
ylist = []
```

```python
# initialization function: plot the background of each frame
def init():
  ax.add_patch(circle)
  ax.add_patch(rect)
  line.set_data([], [])
  return (line,)
```

```python
# animation function. This is called sequentially
def animate(i):
    x = np.random.uniform(-1,1,size=1)
    y = np.random.uniform(-1,1,size=1)
    xlist.append(x)
    ylist.append(y)

    line.set_data(xlist, ylist)
    return (line,)
```

```python
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=500, interval=20, blit=True)
```

```python
HTML(anim.to_html5_video())
```
